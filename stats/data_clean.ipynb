### filter all the idioms that contains a polyphone char

```python
with open('./多音字表.json', 'r', encoding='utf-8') as file:
    dict_polyphones = json.load(file)
```

```python
def contain_polyphone(word):
    for char in word:
        if char in dict_polyphones.keys():
            return True
    return False
```

```python
dict_monophone_idioms = {}
dict_polyphone_idioms = {}
is_polyphone = False
with open(path_pure_idiom_txt, 'r', encoding='utf-8') as file:
    for line in file:
        word = line.strip()
        if (contain_polyphone(word)): 
            dict_polyphone_idioms[word] = ' '.join(py.lazy_pinyin(word, style=py.STYLE_TONE3))
        else:
            dict_monophone_idioms[word] = ' '.join(py.lazy_pinyin(list(word), style=py.STYLE_TONE3)) 
        
```

### correct 多音成语表.json using [polyphones.json](../data/polyphones.json)

```python
def dict_diff(dict1, dict2):
    diff = {}
    miss = {}
    correct_count = 0
    for (key,value) in dict1.items():
        if (not key in dict2.keys()):
            miss[key] = value
        else:
            value2 = dict2[key]
            if (value != value2):
                diff[key] = [value,value2]
            else:
                correct_count += 1
    return diff, miss, correct_count 
```

```python
with open(path_polyphones_json, 'r', encoding='utf-8') as file:
    data = json.load(file)
```
